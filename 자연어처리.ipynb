```python
from konlpy.tag import Okt
```

```python
보고싶다 = """
오늘 네가 보고싶다
널 다시 품에 안아보고 싶다
오늘 네가 난 생각난다
너랑 같이 산책하던 그곳에 서있다
너의 체온이 기억난다
따뜻하게 내 쉬던 숨소리 들려
오늘 네가 온 것 같아
우리 같이 잠들던 벤치에 기대니
시간이 흘러흘러 다시 만날 순 없지
그래도 보고싶다 널 그리운 내 사랑아
오늘 네가 정말 보고싶다
너를 다시 내 품에 안아보고 싶다
오늘 네가 난 생각난다
너를 쓰다듬던 손이 너를 기억한다
니가 보고싶다
"""
```

```python
okt = Okt()
word_token = okt.morphs(보고싶다)
word_token
# '\n': 줄바꿈
```

```python
word_token_nouns = okt.nouns(보고싶다)
word_token_nouns
```

```python
# 왜 불용어로 선택했는가를 보고서에 잘 기제하기
# 컨텐츠상 피해야해서 불용어로 지정하기도 필요나 의미가 없어서 지정하기도
# 뻔한 분석가가 되고 싶지 않으면 어떻게 분석할지 아이디어 내서. 템플릿대로 따라 분석 X 
my_stop_words_ko = ['숨소리','기대니', '가', '랑','\n']

filtered_token = []
for word in word_token:
    if word not in my_stop_words_ko:
        filtered_token.append(word)
    # if "보고싶다" == word: # 비일반적 방법. 만약 보고싶다라는 단어가 나온다면 이렇게 하겠다 ~를 실행
    #     filtered_token.append("보고") # word가 보고싶다라면 쪼개서 단어장에다가 만들어주겠다 하는 내용
    #     filtered_token.append("싶다")
filtered_token
# filtered_token.remove("보고싶다")
# filtered_token
```

```python
filtered_token_over2 = []
for word in word_token:
    if len(word)>=2:
        filtered_token_over2.append(word)
filtered_token_over2
```

## 글자를 숫자로 번역

```python
tok.texts_to_sequences(filtered_token_over2)
```

```python
# vocab_size = 텍스트 데이터의 전체 단어 집합의 크기
# 정렬을 하는 이유: vocab_size를 정해서 안쓰는 단어 버려야할 건 버리기 위해. 
# 단어수에 따라 모델 성능이 달라진다. (분석 시간에도 영향을 미친다. 불필요한 시간을 소요할 필요가 없다.)
vocab_size = 10
tok_vocab_10 = Tokenizer(num_words = vocab_size+1) # OOV
tok_vocab_10.fit_on_texts(filtered_token_over2)
tok_vocab_10.word_index
```

```python
words_freq_cut = [word for word, index in tok_vocab_10.word_index.items()
                 if index >= vocab_size+1]   # 불용어 자르기 위한 코드
words_freq_cut
```

```python
# words_freq_cut 활용
for word in words_freq_cut:
    del tok_vocab_10.word_index[word]
    del tok_vocab_10.word_counts[word]
print(tok_vocab_10.word_index)
print(tok_vocab_10.word_counts)
print(tok_vocab_10.texts_to_sequences(filtered_token_over2))
```

# 코사인 유사도

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```python
def cos_sim(A, B):
    return dot(A, B)/(norm(A) * norm(B))
```

```python
"""
문서 1: 나는 치킨 좋아요
문서 2: 나는 피자 좋아요
문서 3: 나는 피자 좋아요 나는 피자 좋아요

       나는 치킨 피자 좋아요 -> 우리의 vocabulary
문서 1: 1    1   0     1
문서 2: 1    0   1     1
문서 3: 2    0   2     2 
      이런 상태의 metrix 행렬 구성할 수 있음
"""

# 이런 뉴스 기사에 어떤 숫자적 처리를 할 것인가
```

```python
doc1 = np.array([1,1,0,1])
doc2 = np.array([1,0,1,1])
doc3 = np.array([2,0,2,2])
```

```python
# 문서 1: 나는 치킨 좋아요
# 문서 2: 나는 피자 좋아요
# 문서 3: 나는 피자 좋아요 나는 피자 좋아요

cos_sim(doc1, doc2), cos_sim(doc1, doc3), cos_sim(doc2, doc3) 
# 방향이 2/3는 비슷하다고 봐서 이정도 수치가 나온 것
# 같은 문장 반복. 방향이 같다. 강도의 차이(doc3가 강도가 2배 강할뿐) => 같은 강도이기에 1이 나옴
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity([doc1],[doc2]))
print(cosine_similarity([doc1],[doc3]))
print(cosine_similarity([doc2],[doc3]))
```

# 유클리디안(유클리드 거리) 유사도

```python
def eu_dist(A, B):
    return np.sqrt(np.sum((A-B)**2))
```

```python
# 문서 1: 나는 치킨 좋아요
# 문서 2: 나는 피자 좋아요
# 문서 3: 나는 피자 좋아요 나는 피자 좋아요

eu_dist(doc1, doc2), eu_dist(doc1, doc3), eu_dist(doc2, doc3) 
```

```python
from sklearn.metrics.pairwise import euclidean_distances
print(euclidean_distances([doc1],[doc2]))
print(euclidean_distances([doc1],[doc3]))
print(euclidean_distances([doc2],[doc3]))
```

# 영화 데이터 모델

```python
df_movie = pd.read_csv("movies_metadata.csv", low_memory=False)
df_movie.info
df_movie
```

```python
data = df_movie[['title', "overview"]]
data
```

```python
data.info()
# 45460 - 44512 = 948개의 overview none 값
```

```python
data = data.dropna()
data.info()
data = data[:20000]
```

```python
tfidf = TfidfVectorizer(stop_words="english")
tfidf_dtm = tfidf.fit_transform(data["overview"])

# 문서의 개수 x 단어 7만개정도 동원이 됐다.
```

```python
# 코사인 유사도 기반
cos_sim_res = cosine_similarity(tfidf_dtm, tfidf_dtm)
```

```python
title_to_index = dict(zip(data["title"], data.index))
# key로 들어갈 부분은 title
title_to_index
```

```python
title_to_index = dict(zip(data["title"], data.index))
title_to_index
```

```python
def get_recommendation(title,n):
    idx = title_to_index[title]
    sim_scores = list(enumerate(cos_sim_res[idx]))
    sim_scores = sorted(sim_scores,key=lambda x : x[1], reverse=True) #key값으로 정렬
    sim_scores_n = sim_scores[1:n+1]
    movie_idx = [movie_dic[0] for movie_dic in sim_scores_n]
    return data["title"].iloc[movie_idx]
```

```python
get_recommendation("Toy Story", 3)
```
